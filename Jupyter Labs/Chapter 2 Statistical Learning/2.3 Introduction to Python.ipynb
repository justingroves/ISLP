## 2.3.3 Introduction to Numerical Python

In Python, we can make use of function *libraries* or *packages*, which are collections of *modules* that are not necessarily included in the base Python distribution. The package we will use in this section is `numpy`, which is an appreviation for *numerical Python*.

First we must `import` the package. Note that we alias the package as `np`.

```python
import numpy as np
```

In `numpy`, an *array* is a generic term for a multidimensional set of numbers. We can use `np.array()` to define `x` and `y`, which are one-dimensional arrays (i.e., vectors).

```python
x = np.array([3, 4, 5])
y = np.array([4, 9, 7])
```

Note now that the `+` operator when used on two `numpy` arrays actually adds the elements of the arrays together pointwise:

```python
x + y
```

We can represent vectors using one-dimensional `numpy` arrays and matrices using two-dimensional `numpy` arrays. It is possible to represent a matrix using `np.matrix()`, but in this book we will use `np.array()` instead as this is the standard and allows for working with $n$-dimensional arrays which will come in handy later.

Next we create a matrix using `np.array()`:

```python
x = np.array([[1,2], [3,4]])
x
```

In Python, objects (such as a `np.array()`) have multiple *attributes*, or associated objects/variables. We can access these using the syntax `<object>.<attribute>`, such as getting the number of dimensions of our array `x`:

```python
x.ndim
```

We can also get the *data type* of `x` using `<object>.dtype`. In the following example, `int64` indicates that `x` contains 64-bit integers.

```python
x.dtype
```

Python is intelligent, and the `numpy` package automatically selected the minimal data type required to store our object. If we had initialized the array with decimals, it would change the data type. In this case, it will contain 64-bit *floating-point numbers* (i.e., real-valued numbers).

```python
np.array([[1,2], [3.0, 4]]).dtype
```

We can see the documentation of `np.array` using our `?`:

```python
np.array?
```

We can actually set the data type during initialization of the array passing a `dtype` argument.

```python
np.array([[1, 2], [3, 4]], float).dtype
```

We can find the number of rows and columns using the `shape` attribute.

```python
x.shape
```

A *method* is a function that is associated with an object. For instance, given a `np.array`, the expressions `x.sum()` will sum all the elements of the array. 

```python
x = np.array([1, 2, 3, 4])
print(x.sum())
```

Note that the `numpy` package has a `sum` method that we can pass the array to, which accomplishes the same goal of adding the elements together.

```python
print(np.sum(x))
```

We can use the `reshape()` method to re-organize the array with different dimmensions based on values we pass into the function.

```python
x = np.array([1, 2, 3, 4, 5, 6])
print('beginning x:\n', x)
x_reshape = x.reshape((2,3))
print('reshaped x:\n', x_reshape)
```

We can see also that `numpy` arrays are specified as a series of *rows*, which is called *row-major ordering* as opposed to *column-major ordering*.

Python also uses 0-based indexing, meaning the $0^{th}$ element of the array is actually the first element of the array.

```python
print(x_reshape[0,0])
```

Similarly we can access the element in the first row and second column:


```python
print(x_reshape[1,2])
```

Note that if we modify the first element of `x_reshape`, then we also modify the first element of `x`!

```python
print('x before we modify x_reshape:\n', x)
print('x_reshape before we modify x_reshape:\n', x_reshape)
x_reshape[0,0]=5
print('x_reshape after we modify its top left element:\n', x_reshape)
print('x after we modify top left element of x_reshape:\n', x)
```

This happened because Python is set both `x` and `x_reshape` to the same location in memory. In Python, variables are actually *pointers* to objects in memory, rather than the object itself. So when we modify one pointer, we tell Python to modify the object in memory and the pointer variables `x` and `x_reshape` are both pointing to that singular object that has been modified.

We've seen that we can modify lists, but what about tuples? Tuples are set with parentheses `()`.

```python
my_tuple = (3, 4, 5)
my_tuple[0] = 2
```

Tuples are not mutable, meaning that they are set after being initialized. Trying to change a tuple throws an *exception*, or error.

For some extra handy attributes of arrays we have:
- `shape` : returns the array's dimensions
- `ndim`: returns the number of dimensions
- `T`: the transpose of the array

```python
x_reshape.shape, x_reshape.ndim, x_reshape.T
```

Note that the number of dimensions is returned as a tuple, and the output of the previous line is also a tuple.

Throughout this book we will want to generate random data. We can use the `np.random.normal()` function to generate a vector of random normal variables. Let's check the inputs of the function:

```python
np.random.normal?
```

The first line is referred to as the *signature* line, and tells us that the function takes in three *keyword* arguments, meaning that the arguments `loc`, `scale`, and `size` can be passed into the function in any order by using their names. There are also *positional* arguments that do not have keywords associated with them and must be passed into the function in a proper order. Reading the notation we see that
- `loc` : the mean of the distribution
- `scale` : the standard deviation of the distribution
- `size` : the dimension of the vector returned 

We can use these keywords to generate 50 independent random variables sampled from a $N(0,1)$ distribution.

```python
x = np.random.normal(size=50)
x
```

We can create another array `y` by adding an independent $N(50,1)$ random variable to each element of x.

```python
y = x + np.random.normal(loc=50, scale=1, size=50)
```

We can use the `numpy` method `np.corrcoef()` to compute the correlation matrix between `x` and `y`. The off-diagonal elements give the correlation between `x` and `y`.

```python
np.corrcoef(x,y)
```

Note that since these are sampling from a random distribution, if we re-run this notebook we will get different numbers. WE can make sure that we don't get different numbers by setting the *random seed* of `numpy` using the function `np.random.default_rng()`, which essentially has all the random number generating methods found in `np.random`. Hency, to generate normal data we can use `rng.normal()`.

```python
rng = np.random.default_rng(1303)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng(1303)
print(rng2.normal(scale=5, size=2))
```

We will make use of `np.random.default_rng()` to set the seed and try to get our results to match those that were created in the book. However, due to package updates, there may still be some discrepancies. 

We can also use the following `numpy` functions:
- `np.mean()` - computes the mean of arrays
- `np.var()` - computes the variance of arrays
- `np.std()` - computes the standard deviation of arrays

```python
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
print(np.mean(y), y.mean())
```

```python
print(np.var(y), y.var(), np.mean((y-y.mean())**2))
```

Note that by default `np.var()` divides by the sample size $n$ rather than $n-1$. In other words, `numpy` is using the formula

$$ \sigma^2 = \sum_{i=1}^n \frac{(\bar{x} - x_i)^2}{n} $$

which is the formula for the *population standard deviation* as opposed to

$$ s^2 = \sum_{i=1}^n \frac{(\bar{x} - x_i)^2}{n-1} $$

which is the formula for the *sample standard deviation*. Note that dividing by $n-1$ for the sample standard deviation provides a better estimate of the population standard deviation when not all data points are known. Essentially this means that `numpy` is assuming by default that all data points of the population are provided.

We can change this though by using the `ddof = 1` argument with `np.var`.

```python
np.var?
```

```python
print(np.sqrt(np.var(y)), np.std(y))
```

We can also apply the `np.mean()`, `np.var()`, and `np.std()` functions to the rows and columns of a matrix. We start by costructing a $10 \times 3$ matrix of $N(0,1)$ random variables, and consider computing its row sums.

```python
X = rng.standard_normal((10, 3))
X
```

If we want to take the mean of each of the rows, we must specify in the `mean()` method the axis we are using. Otherwise we will end up taking the mean of all values in the array.

```python
print(X.mean(axis=0))
```

We can also drop the keyword `axis` and use the fact that the `axis` is the first positional argument in the method.

```python
X.mean(0)
```

## 2.3.4 Graphics

For graphics we will primarily make use of the `matplotlib` library, and namely the `matplotlib.pyplot` sub-package. We will use the `subplots()` method to create a figure and and axes onto which the figure will lie. 

In `matplotlib`, a plot consists of a *figure* and one or more *axes*. The figure is the blank canvas on which the plots will be displayed, and the *axes* contains information about each plot, such as information about the $x$- and $y$-axis.

The `subplots()` function allows for specifying a grid of figures and axes so we can specify multiple plots in the same image.

```python
from matplotlib.pyplot import subplots
fig, ax = subplots(figsize=(8, 8))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x, y);
```

Unpacking is the process of taking a tuple and storing it in separate variables, such as storing the results of the tuple given by `subplots(figsize=(8,8))` in the two variables `fig` and `ax`. Unpacking is preferred to the equivalent, but more convoluted, process.

```python
output = subplots(figsize=(8,8))
fig = output[0]
ax = output[1]
```

Note that in our previous output, we had a lineplot and not a scatterplot. To create a scatterplot, we have to attach an additional argument to `ax.plot`. 

```python
fig, ax = subplots(figsize=(8,8))
ax.plot(x, y, 'o');
```

Alternatively, we could have used the `ax.scatter()` function as shown:

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(x, y, marker='o');
```

Note that above putting the semicolon `;` at the end of the last line prevents the function from printing the text to the notebook. However, it does not prevent the image from being output. Omitting the training semi-colon we get the following additional text:

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(x, y, marker='o')
```

Moving forward we will use trailing semicolons whenever the text output is not necessary to the discussion at hand.

We can also label our plot as so:

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(x,y, marker='o')
ax.set_xlabel("this is the x-axis")
ax.set_ylabel("this is the y-label")
ax.set_title("Plot of X vs Y");
```

The benefit of the `fig, ax` approach is that it allows us access to the figure itself, and we can make changes such as changing the size from `(8,8)` to `(12,3)`.

```python
fig.set_size_inches(12,3)
fig
```

In the situation that we want to create multiple plots within a figure we can achieve this by passing additional arguments to `subplots()`. Viewing the figure as a grid, we can pass in the number of rows and number of columns of plots. We may want to make sure that all of the figures have the same $x$-axis, which we can do using the parameter `sharx=True`. When using a series of plots, we can capture all the axes using the following syntax.

```python
fig, axes = subplots(nrows=2,
                     ncols=3,
                     figsize=(15,5))
```

We can now control the icon used for each of the different plots when plotting points:

```python
axes[0,1].plot(x,y, 'o')
axes[1,2].scatter(x, y, marker='+')
fig
```

To save our output, we can use the `savefig()` method. The argument `dpi` is the number of dots per inch, used to determine how large the image will be in pixels.

```python
fig.savefig("Figure.png", dpi=400)
fig.savefig("Figure.pdf", dpi=200)
```
